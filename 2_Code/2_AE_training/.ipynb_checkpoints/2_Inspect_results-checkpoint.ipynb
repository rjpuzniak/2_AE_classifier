```python
# Importing libraries
import torchio as tio
import glob
import numpy as np
import random
import os

from collections import OrderedDict
from pathlib import Path

from tqdm import tqdm
import time

import torchio as tio
from torchio.transforms import (RescaleIntensity,RandomFlip,Compose, HistogramStandardization, RandomAffine, RandomNoise, ToCanonical)

from torch.utils.data import DataLoader
import torch
import torch.nn as nn

import matplotlib.pyplot as plt

import pickle
```

```python
# Load the dictionary with study design
with open('study_design.pkl', 'rb') as f:
    design = pickle.load(f)
```

```python
train_groups=['ABIDE','Athletes','HCP','COBRE','Leipzig']
test_dev_groups=['UoN']
test_groups=['CHIASM']

test_design = {}

# Initialize keys
for key in design.keys():
    test_design[key]={}
    
for dataset in train_groups:
    test_design['train'][dataset]={}
    test_design['dev_train'][dataset]={}
    
for dataset in test_dev_groups:
    test_design['dev_test'][dataset]={}    
    
for dataset in test_groups:
    test_design['test'][dataset]={}
    
# Fill the frame of test_design with data
for key in design.keys():
    
    for subject in design[key]['all']:
        
        # Extract path
        path = os.path.dirname(design[key]['all'][subject]['brain'])
        split_path = os.path.normpath(path).split(os.path.sep)
        sub = split_path[-1]
        dataset = split_path[-2]
               
        # Create dictionary       
        files={}
        files['chiasm']=path+'/chiasm.nii.gz'
        files['chiasm_resampled']=path+'/chiasm_sampling.nii.gz'
        files['t1w']=path+'/t1w_1mm_iso.nii.gz'
        
        test_design[key][dataset][sub]=files
    
```

```python
# Save the dictionary
#with open('study_test_design.pkl', 'wb') as f:
#    pickle.dump(test_design, f)

# Load the dictionary
with open('study_test_design.pkl', 'rb') as f:
    test_design = pickle.load(f)
```

```python
# Dictionary with images
subjects_chiasms_list={}

for group in test_design.keys():
    subjects_chiasms_list[group]={}
    
    for dataset in test_design[group].keys():
        
        if dataset=='COBRE':
            continue
                        
        subjects_chiasms_list[group][dataset]={}
        
        gen = (sub for sub in test_design[group][dataset].keys() if os.path.isfile(test_design[group][dataset][sub]['chiasm_resampled']))
          
        subjects_chiasms_list[group][dataset] = [tio.Subject(chiasm=tio.Image(test_design[group][dataset][sub]['chiasm'], type=tio.INTENSITY), chiasm_resampled = tio.Image(test_design[group][dataset][sub]['chiasm_resampled'], type=tio.INTENSITY)) for sub in  gen]     
```

```python
# Transformations
rescale = RescaleIntensity((0,1))
```

```python
# Dataset containing only chiasm images
dataset_chiasm={}

for group in subjects_chiasms_list.keys():
    dataset_chiasm[group]={}
    for dataset in subjects_chiasms_list[group].keys():
        dataset_chiasm[group][dataset]=tio.SubjectsDataset(subjects_chiasms_list[group][dataset], transform = rescale)
        
```

```python
# Dataloader
dataloader_chiasm={}

for group in dataset_chiasm.keys():
    dataloader_chiasm[group]={}
    for dataset in dataset_chiasm[group].keys():
        dataloader_chiasm[group][dataset] = DataLoader(dataset=dataset_chiasm[group][dataset], batch_size=10, shuffle=True, num_workers=8)
```

```python
# Create dictionary that will store all results
results={}

architectures = [x[0] for x in os.walk('../../1_Data/2_Trained_AE/')]

for architecture in architectures[1:]:
    results['Encoder_'+os.path.basename(architecture)]={}

results['Resampling']={}

for method in results.keys():
    for group in dataloader_chiasm.keys():
        results[method][group]={}
        for dataset in dataloader_chiasm[group].keys():
            results[method][group][dataset]={}
```

```python
# Evaluate all the approaches, calculate average MSE loss per image and save it to dictionary
loss_function = nn.MSELoss(reduction='sum')
```

```python
# Cropped U-Net copied from Overfitting Model

class UNet(nn.Module):
    
    def __init__(self, in_channels=1, out_channels=1, init_features=10, scaling=2):
        super(UNet, self).__init__()
                
        # Encoding layers
        self.encoder1 = self.unet_block(in_channels, init_features, "enc1")
        self.pool1 = nn.AvgPool3d(kernel_size=2, stride=2, padding=0)
        self.encoder2 = self.unet_block(init_features, init_features*scaling, name='enc2')
        self.pool2 = nn.AvgPool3d(kernel_size=2, stride=2, padding=0)

        # Bottleneck layer
        self.bottleneck = self.unet_block(init_features*scaling, init_features*scaling**2, name='bottleneck')
        
        # Decoding layers (where merge with prevois encoding layers occurs)        
        self.upconv2 = nn.ConvTranspose3d(init_features*scaling**2, init_features*scaling, kernel_size=2, stride=2)
        self.decoder2 = self.unet_block(init_features*scaling, init_features*scaling, name='dec2')
                
        self.upconv1 = nn.ConvTranspose3d(init_features*scaling, init_features, kernel_size=2, stride=2)
        self.decoder1 = self.unet_block(init_features, init_features, name='dec1')
        
        # Final convolution - output equals number of output channels
        self.conv = nn.Conv3d(init_features, out_channels, kernel_size=1) 
        
    def forward(self,x):
        
        # Encoding
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool1(enc1))

        # Bottleneck
        bottleneck = self.bottleneck(self.pool2(enc2))

        # Upconvolving, concatenating data from respective encoding phase and executing UNet block
        dec2 = self.upconv2(bottleneck)
        dec2 = self.decoder2(dec2)
        dec1 = self.upconv1(dec2)
        dec1 = self.decoder1(dec1)
        
        out_conv = self.conv(dec1)
        
        return torch.sigmoid(out_conv)
    
    def unet_block(self, in_channels, features, name):
        
        return nn.Sequential(OrderedDict([(name+'conv1',nn.Conv3d(in_channels=in_channels, out_channels=features, kernel_size=3, padding=1, bias=False)),
                             (name+'bnorm1', nn.BatchNorm3d(num_features=features)),
                             (name+'relu1', nn.ReLU(inplace=True)),
                             (name+'conv2', nn.Conv3d(in_channels=features, out_channels=features, kernel_size=3, padding=1, bias=False)),
                             (name+'bnorm2', nn.BatchNorm3d(num_features=features)),
                             (name+'relu2', nn.ReLU(inplace=True))])
                            )

    def output_latent_representations(self,x):
        
        print(x.shape)

        # Encoding
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool1(enc1))

        # Bottleneck
        bottleneck = self.bottleneck(self.pool2(enc2))
        
        print(bottleneck.shape)
        
        return bottleneck
```

```python
# For resampling
for group in dataloader_chiasm.keys():
    for dataset in dataloader_chiasm[group].keys():

        total_loss = 0.0
        
        for batch in dataloader_chiasm[group][dataset]:
            
            total_loss += loss_function(batch['chiasm']['data'],batch['chiasm_resampled']['data'])
            
        loss_per_image = total_loss/len(dataset_chiasm[group][dataset])
        
        results['Resampling'][group][dataset] = loss_per_image
```

```python
# For autoencoder
for architecture in architectures[1:]:
    
    # Extract the parameters
    params = os.path.basename(architecture)
    
    # Initialize the network and set in to eval mode
    network = UNet(1,1,int(params.split('_')[0]),int(params.split('_')[1]))
    
    network.load_state_dict(torch.load(architecture+'/optimal_weights'))
    
    network.eval()
    
    for group in dataloader_chiasm.keys():
        for dataset in dataloader_chiasm[group].keys():
            
            total_loss = 0.0
            
            for batch in dataloader_chiasm[group][dataset]:
                
                with torch.no_grad():
                    prediction = network(batch['chiasm']['data'])
                
                total_loss += loss_function(batch['chiasm']['data'],prediction)
                
            loss_per_image = total_loss/len(dataset_chiasm[group][dataset])
            
            results['Encoder_'+params][group][dataset] = loss_per_image.detach().numpy()

# Load model, put it in eval, run on input and compare against original


#..............................................................................
```

```python
def load_training(path):
    
    with open(path+'/train_loss.pkl', 'rb') as f:
        train_loss = pickle.load(f)
    
    with open(path+'/dev_train_loss.pkl', 'rb') as f:
        dev_train_loss = pickle.load(f)

    with open(path+'/dev_test_loss.pkl', 'rb') as f:
        dev_test_loss = pickle.load(f)
    
    return train_loss, dev_train_loss, dev_test_loss    
```

```python
plt.figure(figsize=(20, 10), dpi=300)

plt.title('Training');

candidates=['32_1','8_2','16_1','4_2']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(candidates[i])
    
    train_loss, dev_train_loss, dev_test_loss = load_training('../../1_Data/2_Trained_AE/'+candidates[i])
    
    plt.plot(train_loss, label='train')
    plt.plot(dev_train_loss, label='dev_train')
    plt.plot(dev_test_loss, label='dev_test')
    
    plt.ylim([0,0.008])
    
    plt.xlabel('Epoch')
    plt.ylabel('Avg MSE loss per voxel')
    
    plt.legend()
```
