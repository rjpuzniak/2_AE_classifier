```python
# function that takes loader's batch and outputs corresponding targets - 1 for albinism, 0 for controls
def return_class(batch):
    return [1 if 'ALB'==os.path.basename(os.path.dirname(path))[:3] else 0 for path in batch['mask']['path']]
```

```python
# Function - take network, return the size of bottleneck
def size_of_representation(name,parameters):
       
    argument=(tuple(map(int,parameters.split('_'))))
    
    network=eval(name)(*argument)
    
    output=network(torch.randn(1,1,24,24,8))
    dims = output.detach().numpy().shape
    return dims[0]*dims[1]*dims[2]*dims[3]*dims[4]
```

```python
# function that takes the pretrained network, loads optimal weights, freezes weights, removes decoding layer and replaces those with custom classifying layers
def create_classifier_from_ae(network_arch, params, k_fold, fc_layer_size):
    
    # Change string with parameters into tuple
    argument=(tuple(map(int,params.split('_'))))
    
    # Create an instance of network
    network=eval(network_arch)(*argument)
    
    # Load optimal parameters from previous autoencoder training
    network.load_state_dict(torch.load('../0_Data/3_Autoencoders/'+network_arch+'/'+params+'/'+k_fold+'/optimal_weights.pt'))    
   
    # Calculate dimensions of representation
    rep_size=size_of_representation('encode_'+network_arch,params)
    
    # Remove decoding layer
    layers=[]
    layers.append(nn.Flatten())
    layers.append(nn.Linear(rep_size, fc_layer_size))
    layers.append(nn.ReLU())
    layers.append(nn.Dropout(0.2)) # previously 0.2
    layers.append(nn.Linear(fc_layer_size, 1))

    classifying_layer = nn.Sequential(*layers)
    
    network.decoder1 = classifying_layer

    # Adjusting one layer network
    if network_arch=='one_layer':
        
        # Freeze weights
        for param in network.encoder1.parameters():
            param.requires_grad = False

        for param in network.pool1.parameters():
            param.requires_grad = False
    
    if network_arch=='two_layers':
        
        # Freeze weights
        for param in network.encoder1.parameters():
            param.requires_grad = False

        for param in network.pool1.parameters():
            param.requires_grad = False
            
        for param in network.encoder2.parameters():
            param.requires_grad = False
            
        for param in network.pool2.parameters():
            param.requires_grad = False
            
        network.decoder2 = nn.Sequential()
        network.decoder3 = nn.Sequential()
        
    if network_arch=='three_layers':
        
        # Freeze weights
        for param in network.encoder1.parameters():
            param.requires_grad = False

        for param in network.pool1.parameters():
            param.requires_grad = False
            
        for param in network.encoder2.parameters():
            param.requires_grad = False
            
        for param in network.pool2.parameters():
            param.requires_grad = False
            
        for param in network.encoder3.parameters():
            param.requires_grad = False
            
        for param in network.pool3.parameters():
            param.requires_grad = False
            
        network.decoder2 = nn.Sequential()
        network.decoder3 = nn.Sequential()
        network.decoder4 = nn.Sequential()
            
    # Modify output activation function
    network.output = nn.Sigmoid()
    
    return network 
```

```python
# function that takes the pretrained network, loads optimal weights, freezes weights, removes decoding layer and replaces those with custom classifying layers
def create_2_layer_classifier_from_ae(network_arch, params, k_fold, fc_layer_size):
    
    # Change string with parameters into tuple
    argument=(tuple(map(int,params.split('_'))))
    
    # Create an instance of network
    network=eval(network_arch)(*argument)
    
    # Load optimal parameters from previous autoencoder training
    #network.load_state_dict(torch.load('../0_Data/3_Autoencoders/'+network_arch+'/'+params+'/'+k_fold+'/optimal_weights.pt'))    
   
    # Calculate dimensions of representation
    rep_size=size_of_representation('encode_'+network_arch,params)
    
    # Remove decoding layer
    
    network.decoder1 = nn.Flatten()
    
    layers=[]
    layers.append(nn.Linear(rep_size, fc_layer_size))
    layers.append(nn.ReLU())
    layers.append(nn.Dropout(0.2)) # previously 0.2
    
    layers.append(nn.Linear(fc_layer_size, 2*fc_layer_size))
    layers.append(nn.ReLU())
    layers.append(nn.Dropout(0.2)) # previously 0.2
    classifying_layer1 = nn.Sequential(*layers)
    
    classifying_layer1 = nn.Sequential(*layers)
    
    network.decoder2 = classifying_layer1
    
    layers=[]
    layers.append(nn.Linear(2*fc_layer_size, fc_layer_size))
    layers.append(nn.ReLU())
    layers.append(nn.Dropout(0.2)) # previously 0.2
    layers.append(nn.Linear(fc_layer_size,1))
    classifying_layer2 = nn.Sequential(*layers)
    
    network.decoder3 = classifying_layer2
    
    # Freeze weights
    for param in network.encoder1.parameters():
        param.requires_grad = False

    for param in network.pool1.parameters():
        param.requires_grad = False

    for param in network.encoder2.parameters():
        param.requires_grad = False

    for param in network.pool2.parameters():
        param.requires_grad = False

    for param in network.encoder3.parameters():
        param.requires_grad = False

    for param in network.pool3.parameters():
        param.requires_grad = False
            
    # Modify output activation function
    network.output = nn.Sigmoid()
    
    return network 
```

```python
# Function returning parameters of the best AE (for given architecture)
def optimal_parameters(dictionary, arch):
    return max(dictionary[arch], key=lambda k: dictionary[arch][k]['dsc_mean'])
```

```python
def train_classifier(network, dataloader, optimizer, criterion, n_epochs, device, save_path):
    
    track_validation_loss = []
    track_training_loss = []
    
    validation_loss_min = np.Inf
    
    for epoch in tqdm(range(1, n_epochs+1)):
        
        training_loss = []
        validation_loss = []
        
        # Train
        network.train()
        for batch_idx, batch in enumerate(dataloader['train']):
            
            data=batch['mask']['data']
            target = torch.tensor(return_class(batch)).unsqueeze(1)
            
            optimizer.zero_grad()
            
            output = network(data)
            
            loss = criterion(output.float(),target.float())
            
            loss.backward()
            optimizer.step()
            
            training_loss.append(loss.mean().item())
            
        # Validate
        network.eval()
        for batch_idx, batch in enumerate(dataloader['valid']):
            
            data = batch['mask']['data']
            target = torch.tensor(return_class(batch)).unsqueeze(1)
            
            with torch.no_grad():
                output = network(data)
                loss = criterion(output.float(),target.float())
                
                validation_loss.append(loss.mean().item())
                
        # Calculate average loss per batch
        training_loss = sum(training_loss)/len(training_loss)
        validation_loss = sum(validation_loss)/len(validation_loss)
        
        # Update optimal model
        if validation_loss < validation_loss_min:
            validation_loss_min = validation_loss
            torch.save(network.state_dict(),save_path+'optimal_weights.pt')
            
        # Add values to the list
        track_training_loss.append(training_loss)
        track_validation_loss.append(validation_loss)
        
    return track_training_loss, track_validation_loss
```

```python
# Train and evaluate classification
network_types= ['three_layers']

# FC layer sizes
fc_layer_sizes=[256] # best for 1024

criterion = nn.BCELoss()

n_epochs = 10000

# Try setting CUDA if possible
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu") 

# Load the dictionary
with open('Auxillary_files/evaluation_dictionary.pkl', 'rb') as f:
        evaluation_dictionary = pickle.load(f)

for network_type in network_types:
        
    params = optimal_parameters(evaluation_dictionary, network_type)
    argument=(tuple(map(int,params.split('_'))))
    
    for k_fold in range(12):
        
        print('12-fold validation, stack {}'.format(k_fold+1))
        
        # Create instance of the network with optimal parameters
        base_network=eval(network_type)(*argument)
        
        # Load the dataloader
        dataloader = torch.load('Auxillary_files/Dataloaders/dataloader_'+str(k_fold+1)+'.pth')
        
        for fc_layer_size in fc_layer_sizes:
            
            # Modify the network
            network = create_classifier_from_ae(network_type, params, str(k_fold+1), fc_layer_size)
            #network = create_2_layer_classifier_from_ae(network_type,params,str(k_fold+1), fc_layer_size)
            
            # Create an optimizer
            optimizer = torch.optim.Adam(params=network.parameters(), lr=0.00001) # best for 0.0001 , but also 0.000005
            
            # Define savepath and create folders if necessary
            savepath='../0_Data/4_Classifiers/'+network_type+'_'+params+'/'+str(fc_layer_size)+'/'+str(k_fold+1)+'/'
            if not os.path.exists(savepath):
                os.makedirs(savepath)
        
            # Train
            track_training_loss, track_validation_loss = train_classifier(network,dataloader,optimizer, criterion, n_epochs, device, savepath)
        
            # Print training results
            plt.plot(track_training_loss)
            plt.plot(track_validation_loss)
            plt.legend(['training','validation'])
            plt.show()
            
            # Save tracked results
            with open(savepath+'track_training.txt','wb') as file:
                pickle.dump(track_training_loss,file)

            with open(savepath+'track_validation.txt','wb') as file:
                pickle.dump(track_validation_loss,file)
                
            # Load the optimal model
            optimal_network = create_classifier_from_ae(network_type, params, str(k_fold+1), fc_layer_size)            
            optimal_network.load_state_dict(torch.load(savepath+'optimal_weights.pt'))    
            optimal_network.eval()
            
            # Evaluate the test dataset on the optimal model
            results = []
            
            for batch_idx, batch in enumerate(dataloader['test']):
                
                data = batch['mask']['data']
                target = torch.tensor(return_class(batch))
                
                with torch.no_grad():
                    prediction = optimal_network(data)
                    
                results.append((batch['mask']['path'],target.item(),prediction.detach().numpy()))
                
            # Print and save results
            
            with open(savepath+'performance_on_test_data.txt','wb') as results_file:
                pickle.dump(results, results_file)
```
